```python
plmode = df['painloc'].mode()[0]
df['painloc'] = df['painloc'].fillna(plmode)
pemode = df['painexer'].mode()[0]
df['painexer'] = df['painexer'].fillna(pemode)

df.loc[df['trestbps'] < 100, 'trestbps'] = 100
trestbps_mean = df.loc[:, 'trestbps'].mean()
df['trestbps'] = df['trestbps'].fillna(trestbps_mean)

df.loc[df['oldpeak'] < 0, 'oldpeak'] = 0
df.loc[df['oldpeak'] > 4, 'oldpeak'] = 4
oldpeak_mean = df.loc[:, 'oldpeak'].mean()
df['oldpeak'] = df['oldpeak'].fillna(oldpeak_mean)

thaldur_mean = df.loc[:, 'thaldur'].mean()
thalach_mean = df.loc[:, 'thalach'].mean()
df['thaldur'] = df['thaldur'].fillna(thaldur_mean)
df['thalach'] = df['thalach'].fillna(thalach_mean)

ilist = ['fbs', 'prop', 'nitr', 'pro', 'diuretic', 'exang']
for i in ilist:
    df.loc[df[i] > 1, i] = 1
    mode = df[i].mode()[0]
    df[i] = df[i].fillna(mode)

slope_mean = df.loc[:, 'slope'].mean()
df['slope'] = df['slope'].fillna(slope_mean)
df
```

```python
import requests
from bs4 import BeautifulSoup
req = requests.get('https://www.abs.gov.au/statistics/health/health-conditions-and-risks/smoking-and-vaping/latest-release')
soup = BeautifulSoup(req.content, 'html.parser')
content = soup.find_all('div', class_='chart-table-group')
next = content[2].find_all('div', class_='abs-file-field-chart-wrapper panel-wrapper table-wrapper no-footer') # this gets the right table
bodies = next[0].find('tbody')
df['age'] = pd.to_numeric(df['age'])
df['smoke_1'] = df['smoke'].copy()
for r in bodies.find_all('tr'):
  range = r.find('th').text.split('–')
  #print(range)
  value = float(r.find_all('td')[9].text) # 9 has the data from 2022
  #print(value)
  if len(range) != 1:
    low_bound = int(range[0])
    high_bound = int(range[1])
    df.loc[(np.isnan(df['smoke_1'])) & (df['age'] >= low_bound) & (df['age'] <= high_bound), 'smoke_1'] = value / 100.0
  else:
    df.loc[(np.isnan(df['smoke_1'])) & (df['age'] >= 75), 'smoke_1'] = value / 100.0
df['smoke_1']
```

```python
df['smoke_2'] = df['smoke'].copy()
df['sex'] = pd.to_numeric(df['sex'])

df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 18) & (df['age'] <= 24) & (df['sex'] == 0), 'smoke_2'] = 0.048
df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 18) & (df['age'] <= 24) & (df['sex'] == 1), 'smoke_2'] = 0.048 * 1.32

df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 25) & (df['age'] <= 44) & (df['sex'] == 0), 'smoke_2'] = 0.125
df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 25) & (df['age'] <= 44) & (df['sex'] == 1), 'smoke_2'] = 0.125 * 1.32

df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 45) & (df['age'] <= 64) & (df['sex'] == 0), 'smoke_2'] = 0.151
df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 45) & (df['age'] <= 64) & (df['sex'] == 1), 'smoke_2'] = 0.151 * 1.32

df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 65) & (df['sex'] == 0), 'smoke_2'] = 0.087
df.loc[(np.isnan(df['smoke_2'])) & (df['age'] >= 65) & (df['sex'] == 1), 'smoke_2'] = 0.087 * 1.32
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(["target", "smoke"], axis=1), df["target"], stratify=df["target"], test_size=0.1)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
lin_scores = cross_val_score(lin_model, X_train, y_train, cv=5)
print("Linear CV and accuracy:")
print(lin_scores)
print(lin_model.score(X_train, y_train))

log_model = LogisticRegression(max_iter=2000)
log_model.fit(X_train, y_train)
log_scores = cross_val_score(log_model, X_train, y_train, cv=5)
print("\nLogistic CV and accuracy:")
print(log_scores)
print(log_model.score(X_train, y_train))

knn_model = KNeighborsClassifier(n_neighbors=15) # played with this number, settled on 15
knn_model.fit(X_train, y_train)
knn_y_pred = knn_model.predict(X_test)
knn_scores = cross_val_score(knn_model, X_train, y_train, cv=5)
print("\nKNN CV and accuracy:")
print(knn_scores)
print(accuracy_score(y_test, knn_y_pred))

print("Logistic regression performed the best.")
```
